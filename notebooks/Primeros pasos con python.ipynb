* Los nombres **NO** se pueden inicializar con un numero o un caracter especial.

```python
d_22 = "Si"
d_22
```

```python
## Asignación de valor a una variable
my_var = 88
type(my_var)
```

```python
st = "Suma"
type(st)
```

```python
# Asignar valores a diferentes variables

a = 6
b = 4

# Asignar la multiplicación de 2 variables
c = a * b

c
```

# Listas

```python
# Definir lista
my_list = [2,4,6,8]
my_list
```
